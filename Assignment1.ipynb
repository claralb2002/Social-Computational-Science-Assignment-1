## Part 1 - Web-scraping

```python
from bs4 import BeautifulSoup # A package to work with HTML data
import requests #A package to make HTTP requests
import unicodedata
import re # Used to remove single letters in names
import pandas as pd
```

```python
LINK = "https://ic2s2-2023.org/program" # Get link

r = requests.get(LINK) # Use Requests package

soup = BeautifulSoup(r.content) # Soup it
```

```python
names = soup.find_all("i")
names
```

```python
# Assuming 'names' is a ResultSet (list of elements) returned by find_all()
clean_names = [name.text for name in names]  # Extract text from each element

# Join all names into a single string and split on ',' to handle multiple names
clean_names = [name.strip() for name in ", ".join(clean_names).split(",")]

#Remove "Chair: " (case-insensitive) and strip whitespace
clean_names = [name.replace("Chair: ", "").replace("chair: ", "").strip() for name in clean_names]

#Remove duplicates while maintaining order
seen = set()
clean_names = [name for name in clean_names if not (name in seen or seen.add(name))]


clean_names = [name.strip() for name in ", ".join(clean_names).split(",")]  # Normalize spacing and split names
clean_names = ["".join(c for c in unicodedata.normalize("NFD", name) if not unicodedata.combining(c)) for name in clean_names]  # Remove diacritics (hats)
clean_names = [name.replace("-", " ").replace("."," ") for name in clean_names]  # Replace - with space
clean_names = [re.sub(r'\s+[A-Za-z]\s+', ' ', name) for name in clean_names] # Removes single letters in names

clean_names = set(clean_names) # Remove all exact duplicates

print(clean_names)  # List of cleaned unique names

len(clean_names)
```

```python
df = pd.DataFrame(clean_names) # Creates pandas dataframe

df.to_csv("Authors_2023.csv",index=False) # Save in seperate file
```

### Retrieve Data

```python
import requests
import seaborn as sns
import pandas as pd
from tqdm import tqdm
from time import sleep
```

```python
#filtrer authors to include only those > 5
all_authors = pd.read_csv("author_data2024.csv", header=None) #all authors
all_authors[2] = pd.to_numeric(all_authors[2], errors="coerce") #from string to number

filtered_authors = all_authors[all_authors[2] > 5]  #Remove all below 5
filtered_authors = filtered_authors[filtered_authors[2] < 5000] #Remove all above 5000

df = filtered_authors[0]
df
```

```python
import requests
import pandas as pd
from time import sleep
from tqdm import tqdm
from joblib import Parallel, delayed

BASE_URL = "https://api.openalex.org/works?filter="

# Define a function to fetch data for a single author
def fetch_work_data(author):
    extracted_data = []
    Filter1 = "author.id:"
    Filter2 = ",authors_count:<10,cited_by_count:>10,concepts.id:C126322002|C136264566|C19165224|C17744445,concepts.id:C33923547|C2522767161|C41008148"
    Filter3 = "&per-page=200&select=id,title,publication_year,cited_by_count,concepts,display_name,abstract_inverted_index,authorships&cursor="
    cursor = "*"
    
    while cursor:
        URL = BASE_URL + Filter1 + author + Filter2 + Filter3 + cursor
        
        response = requests.get(URL)
        
        if not response.ok:
            print(f"Error fetching data for {author}: {response.status_code}")
            sleep(1)
            continue
        
        json_data = response.json()
        cursor = json_data["meta"].get("next_cursor")
        
        for work in json_data.get("results", []):
            extracted_data.append({
                "id": work.get("id", ""),
                "publication_year": work.get("publication_year", ""),
                "cited_by_count": work.get("cited_by_count", 0),
                "author_ids": [auth["author"]["id"] for auth in work.get("authorships", [])],
                "title": work.get("title", ""),
                "abstract_inverted_index": work.get("abstract_inverted_index", None)
            })
    
    return extracted_data

# Run parallel processing
num_jobs = -1  
author_list = df 
all_extracted_data = Parallel(n_jobs=num_jobs)(delayed(fetch_work_data)(author) for author in tqdm(df))

# Flatten list of lists
final_data = [item for sublist in all_extracted_data for item in sublist]

# Convert to DataFrame
extracted_data = pd.DataFrame(final_data)
```

# Part 4 - The Network of Computational Social Science

### Part 1 - Network Construction

```python
import pandas as pd
from itertools import combinations
from collections import Counter
```

```python
## 

co_authors = pd.read_csv("papers2024co_authors.csv")
authors = pd.read_csv("papers2024authors.csv")

works = pd.concat([authors,co_authors],ignore_index = True)

#Create DataFrame
df = pd.DataFrame(works, columns=['id', 'publication_year', 'cited_by_count', 'author_ids'])

# Convert author_ids from string to list
df['author_ids'] = df['author_ids'].apply(eval)

# Generate all author pairs and count collaborations
collab_counter = Counter()

for authors in df['author_ids']:
    for pair in combinations(sorted(authors), 2):  # Sort to avoid duplicates (A, B) and (B, A)
        collab_counter[pair] += 1

# Convert Counter to DataFrame
collab_df = pd.DataFrame(collab_counter.items(), columns=['author_pair', 'collaboration_count'])

# Split tuples into separate columns for clarity
collab_df[['author_1', 'author_2']] = pd.DataFrame(collab_df['author_pair'].tolist(), index=collab_df.index)
collab_df.drop(columns=['author_pair'], inplace=True)
collab_df
```

```python
import networkx as nx
import matplotlib.pyplot as plt

#Load datasets
co_authors = pd.read_csv("co_author_data.csv")
authors = pd.read_csv("author_data.csv")

co_authors_papers = pd.read_csv("papers2024co_authors.csv")
authors_papers = pd.read_csv("papers2024authors.csv")

#Combine co_author and author datasets
papers = pd.concat([authors_papers,co_authors_papers],ignore_index = True)
author_data = pd.concat([authors,co_authors],ignore_index = True)

#Create undirected graph
G = nx.Graph()

#Compute total citations per author
papers['author_ids'] = papers['author_ids'].apply(eval)  # Convert from string to list if needed
papers_exploded = papers.explode('author_ids')
author_citations = papers_exploded.groupby('author_ids')['cited_by_count'].sum().to_dict()

#Compute first publication year for each author
author_years = papers.explode('author_ids')[['author_ids', 'publication_year']]
first_pub_years = author_years.groupby('author_ids')['publication_year'].min().todict()

#Add nodes (authors) with attributes
for _, row in author_data.iterrows():
    G.add_node(row["id"], 
               display_name=row['display_name'], 
               country=row.get('country_code', 'Unknown'), 
               first_pub_year=first_pub_years.get(row["id"], "Unknown"), 
               total_citations=author_citations.get(row["id"], 0))

#Add weighted edges (collaborations)
edges = [(row['author_1'], row['author_2'], row['collaborationcount']) for _, row in collab_df.iterrows()]
G.add_weighted_edges_from(edges)
```

```python
import json

#Save graph as JSON
graph_data = nx.node_link_data(G)
with open("author_network.json", "w") as f:
    json.dump(graph_data, f, indent=4)
```

> 2. **Degree Analysis:**
>    - Compute the average, median, mode, minimum, and maximum degree of the nodes. Perform the same analysis for node strength (weighted degree). What do these metrics tell us about the network? __(answer in max 150 words)__



```python
import statistics

# Compute degree (number of connections per node)
degrees = [deg for _, deg in G.degree()]
weighted_degrees = [deg for _, deg in G.degree(weight='weight')]

# Compute statistics
degree_stats = {
    "average_degree": sum(degrees) / len(degrees),
    "median_degree": statistics.median(degrees),
    "mode_degree": statistics.mode(degrees) if len(set(degrees)) > 1 else "No mode",
    "min_degree": min(degrees),
    "max_degree": max(degrees)
}

weighted_degree_stats = {
    "average_weighted_degree": sum(weighted_degrees) / len(weighted_degrees),
    "median_weighted_degree": statistics.median(weighted_degrees),
    "mode_weighted_degree": statistics.mode(weighted_degrees) if len(set(weighted_degrees)) > 1 else "No mode",
    "min_weighted_degree": min(weighted_degrees),
    "max_weighted_degree": max(weighted_degrees)
}

# Print results
print("Degree Statistics:", degree_stats)
print("Weighted Degree Statistics:", weighted_degree_stats)
```

The average degree of 6.63 suggests moderate collaboration, while the median of 5.0 and mode of 4 indicates that most of the authors have a small number of co-authors. However, the maximum degree of 420 reveals that a few of the authors is highly connected and probably central figures in the field. The weighted degree follows a similar pattern, with an average of 14.28 but a maximum of 1517, showing again that some of the authors have a really high collaboration intensity. This analysis shows a few highly influential authors bridge different research clusters, while most authors collaborate within smaller groups.
